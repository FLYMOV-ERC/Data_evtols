## Start point and connect the other ones, plotting route 

```python
df = pd.DataFrame(data)

# Randomly select 4 points from the DataFrame
#points_data = #df.sample(n=20).reset_index(drop=True)
points_data = df.reset_index(drop=True)


points_data
```

```python
# Select rows by position using .iloc
#df_points = points_data.iloc[station_points]

# Create a base map centered around the average coordinates of the starting points
avg_lat = points_data['lat'].mean()
avg_lon = points_data['lon'].mean()
map_folium = folium.Map(location=[avg_lat, avg_lon], zoom_start=10)

# Filter the DataFrame to exclude starting points
#df_station_points = df_points.drop(start_point)



# Add the other points in blue
for index, row in points_data.iloc[:].iterrows():
    folium.Marker(
        location=[row['lat'], row['lon']],
        popup=f"Route: {row['route_name']}, Altitude: {row['max_alt']}m, Number point: {index}",
        icon=folium.Icon(color='blue', icon='info-sign')
    ).add_to(map_folium)

map_folium


```
